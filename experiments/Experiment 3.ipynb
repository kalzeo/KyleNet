## Experiment 3.
##### This experiment is ran using a balanced dataset but includes image augmentation. Image augmentation is only applied to the training data to provide additional training sample variations.
##### Horizontal flips is the only image augmentation technique applied.
##### The hyperparameters & conditions for this experiment are:
- 20 epochs
- 128 batch size
- 1e-4 learning rate
- Balanced Dataset + Dataset Shuffle
- Image Augmentation (horizontal flips)

```python
import tensorflow as tf
from KyleNet import KyleNet
from pathlib import Path

# Get the metadata
META = Path(".").resolve().parent.joinpath("metadata.csv")
```

```python
if tf.test.gpu_device_name():
    print(f"GPU Device: {tf.test.gpu_device_name()}\n")
else:
    print("Running off of CPU\n")
```
