##### Script to visualize the architecture of the experimental model

```python
import visualkeras
from pathlib import Path
from tensorflow.keras.models import load_model
from collections import defaultdict
from tensorflow.python.keras.layers import Dense, Conv2D, Flatten, BatchNormalization, MaxPooling2D

# Load a KyleNet model
model_dir = Path(".").resolve().parent.joinpath("models")
model = load_model(model_dir.joinpath("Experiment 2.h5"))
```

##### View the model summary

```python
model.summary()
```

##### Configure the color of each layer in the model


```python
color_map = defaultdict(dict)
color_map[Conv2D]["fill"] = "orange"
color_map[BatchNormalization]["fill"] = "gray"
color_map[MaxPooling2D]["fill"] = "red"
color_map[Dense]["fill"] = "green"
color_map[Flatten]["fill"] = "teal"
```

##### View the model architecture and output the diagram to a directory

```python
visualkeras.layered_view(model,
                         color_map=color_map,
                         to_file=model_dir.joinpath("model architecture.png"),
                         spacing=85)
```
